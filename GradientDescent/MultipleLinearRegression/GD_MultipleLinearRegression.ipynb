An extension of gradient descent for a single feature.
The training data has multiple features and gradient descent algorithm is implemented for all features

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
b = 785.1811367994083
w = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
```

```python
def predict(X_train, w, b):
    prediction = np.dot(X_train, w) + b
    return prediction
```

```python
# Testing the predict function on a single data point from X_train:
X1 = X_train[0]
X1_prediction = predict(X1, w, b)
print(f"Prediction for X1 is {X1_prediction}")
```

```python
def compute_mse(X_train, y_train, w, b):
    training_data_size = X_train.shape[0]
    cost = 0
    for i in range(training_data_size):
        prediction = predict(X_train[i],w,b)
        cost += (prediction - y_train[i])**2
    cost = cost/ (2*training_data_size)
    return cost
```

```python
# Compute Cost for current w and b:
cost = compute_mse(X_train, y_train, w, b)
print(f"Cost for current w and b: {cost}")
```

```python
def gradient(X_train, y_train, w, b):
  training_data_size, no_of_features = X_train.shape
  w_delta = np.zeros((no_of_features))
  b_delta = 0

  for i in range(training_data_size):
    model_prediction = np.dot(X_train[i],w)+b
    error = model_prediction - y_train[i]
    for j in range(no_of_features):
      w_delta[j] += error*X_train[i,j]
    b_delta += error

  w_delta = w_delta/training_data_size
  b_delta = b_delta/training_data_size

  return w_delta, b_delta
```

```python
print(gradient(X_train, y_train, w, b))
```

```python
def gradient_descent(X_train, y_train, w, b):
    training_data_size, no_of_features = X_train.shape
    alpha = 9.5e-8
    flag = True
    w_delta = np.zeros((no_of_features,))
    b_delta = 0
    iter = 1
    while(flag):
        if iter >500000:
            print(f"reached termination condition, w is {w} and b is {b}")
            break
        w_delta, b_delta = gradient(X_train, y_train, w, b)

        if iter%100==0:
          print(f"iter: {iter} w_delta{w_delta} b:{b}")

        w_temp = w - alpha * w_delta
        b_temp = b - alpha * b_delta
        if (w_temp == w).all() or b_temp==b:
            print(f"reached termination condition, w is {w} and b is {b}")
            break
        w = w_temp
        b = b_temp

        iter+=1

    return w,b


```

```python
w = np.zeros_like(w)
b = 0
w_opt, b_opt= gradient_descent(X_train, y_train, w, b)
```

```python
for i in range(X_train.shape[0]):
  model_prediction = predict(X_train[i], w_opt, b_opt)
  print(f"Model Prediction : {model_prediction :.2f}")
  print(f"Actual Value     : {y_train[i]}")
cost = compute_mse(X_train, y_train, w_opt, b_opt)
print(f"Cost for w and b given by gradient descent: {cost :.2f}")
```
